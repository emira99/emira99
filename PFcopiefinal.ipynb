```python
from skimage.feature import hog
import numpy as np
def hogs(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    dim=(50,50)
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    gray = resized/255
    vector = hog(gray, orientations=12, block_norm='L1', pixels_per_cell=[5, 5], cells_per_block=[2, 2], feature_vector=True, visualize=False, transform_sqrt=True)

    return np.array(vector)
```

```python
##################
import os
import numpy as np
from skimage.feature import hog
from skimage import io

def extract_hog_features(path_image):
    hog_features = []
    for image_path in path_image:
        try:
            # Load the image
            image = io.imread(path_image, as_gray=True)

            # Extract HOG features
            features = hog(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2))

            hog_features.append(features)
        except Exception as e:
            print(f"Error processing {path_image}: {str(e)}")
            hog_features.append(None)

    return hog_features
```

```python
def euclidean_distance(a, b):
    import numpy as np

    dist = np.sqrt(np.sum(np.square(a-b)))
    return dist
```

```python
def compte_dist(v_train,v_test):
    import numpy as np
    list_im_Test= read_file('/content/drive/MyDrive/GeorgeData/test1.txt')
    i=0
    list_dict=dict()
    for im in list_im_Test:
        ##read feats test
        print(i)
        v_testx=v_test[i]
        j=0
        for im_test in list_im_Test:
            v_train1=v_train[j]
            #compute dict
            d=euclidean_distance(v_testx,v_train1)
            print(im,im_test,d)
            #print(im_train)
            #print(d)
            ###create dict
            list_dict[im_test]=d

            j+=1

        ###computing finished for each image in test set
        a = sorted(list_dict.items(), key=lambda x: x[1],reverse=False)
        i+=1
        fsave=im
        fsave=fsave.replace('png','txt')
        sab='/content/drive/MyDrive/GeorgeData/hog_euclidian/'

        with open(sab + fsave, 'w') as output_file:
            print(a, file=output_file)
```

```python
def split_train_test():

    import numpy as np


    featsTrain=read_file('/content/drive/MyDrive/GeorgeData/feats_train_hog.txt')
    featsTest=read_file('/content/drive/MyDrive/GeorgeData/feats_test_hog.txt')
    list_im_Train=read_file('/content/drive/MyDrive/GeorgeData/train1.txt')
    list_im_Test= read_file('/content/drive/MyDrive/GeorgeData/test1.txt')

    X=[]
    j=0
    #parcourir la liste des images train
    for im_train in list_im_Train:
        #im_train contient le nom de l'image a la position j
        #v_train1 contient le vecteur carateristique relatif à l'image à la position j
        v_train1=featsTrain[j]
        #convertir le vecteur du texte en liste (separateur est le point virgule)
        v_trainspl=v_train1.split(',')
        ##j'affiche la longueur de mon vecteur de caractéristique
        #print('longueur feats  ',len(v_trainspl))
        ##convertir v_trainspl en array numpy
        v_trainx=np.asarray(v_trainspl)
        #forcer le type des valeur en float
        v_train=v_trainx.astype(np.float)
        #je remplis une nouvelle liste X avec les vecteur v_train
        X.append(v_train)
        j+=1
    #convertir la liste X en array
    X=np.array(X)###ensemble train


##le meme travail serait repeté avec l'ensemble de test
    i=0
    Y=[]
    for im in list_im_Test:
        print(im)
        ##read feats test

        v_test1=featsTest[i]
        v_testspl=v_test1.split(',')
        v_testx=np.asarray(v_testspl)
        v_test = v_testx.astype(np.float)
        Y.append(v_test)
        i+=1


    Y=np.array(Y)##ensemble de test (test set)
    return X,Y
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#charger les données dans X_train et X_test (features)
X_train,X_test=split_train_test()

print("X_train data shape: ", X_train.shape)
print("X_test data shape: ", X_test.shape)

#*****************normalisation des features************************
scaler = StandardScaler()
# Fit on training set only.
#concatener les feaures pour l'étape de normalisation
alldata=np.concatenate((X_train,X_test), axis=0)
## training
scaler.fit(alldata)

# Apply transform to both the training set and the test set.
X_train_normalised = scaler.transform(X_train)
X_test_normalised = scaler.transform(X_test)
##appliquer le PCA pour réduire la taille des données et selectionné les composantes pertinantes
#concatener les données
alldata_normalised=np.concatenate((X_train_normalised,X_test_normalised), axis=0)
#appeler l'algorithme PCA, nombre composants : 100
pca = PCA(n_components=100) #500
#entrainement du PCA
pca.fit(alldata_normalised)

#transformation des données: data reduction
X_test_selected = pca.transform(X_test_normalised)
```

```python

def normword(word):
    word=word.replace('s-s_cm','s')
    word=word.replace('s-s_mi','s')
    word=word.replace('s-s_pt','s')
    word=word.replace('s-s_sq','s')
    word=word.replace('s-s_qo','s')
    word=word.replace('s_cm','s')
    word=word.replace('s_mi','s')
    word=word.replace('s_pt','s')
    word=word.replace('s_sq','s')
    word=word.replace('s_qo','s')
    #word=word.lower()
    return word
```

```python
def cal_average(num):
    sum_num = 0
    for t in num:
        sum_num = sum_num + t

    avg = sum_num / len(num)
    return avg
```

```python
def get_vector_xception(path):
    # load model
    model = Xception()
    # remove the output layer
    model = Model(inputs=model.inputs, outputs=model.layers[-2].output)
    # load an image from file
    image = load_img(path, target_size=(299, 299))

    # convert the image pixels to a numpy array
    image = img_to_array(image)

    # reshape data for the model
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

    # prepare the image for the resnet model
    image = preprocess_input(image)

    # get extracted features
    features = model.predict(image)
    features=np.array(features.flatten())

    # save to file
    return features
```

```python
def loadmodel():
    from transformers import ViTFeatureExtractor, ViTModel

    feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')
    model = ViTModel.from_pretrained('google/vit-base-patch16-224-in21k')
    return model,feature_extractor
```

```python
def get_vector_transformer(path):
    # load model
    model,feature_extractor=loadmodel()
    #print('le modèle: ', model)
    #print('le feature extracytor: ', feature_extractor)
    original_image = Image.open(path)
    newsize = (100, 100)
    original_image = original_image.resize(newsize)
    #print('image originale: ', original_image)
    rgb_image = original_image.convert('RGB')
    #print('image RGB: ', rgb_image)

    inputs = feature_extractor(images=rgb_image, return_tensors="pt")
    #print('inputs: ', inputs)
    outputs = model(**inputs)
    #print('outputs: ', outputs)
    last_hidden_states = outputs.last_hidden_state
    last_hidden_states=last_hidden_states.flatten()
    #print('last_hidden_states : ' ,last_hidden_states)
    feats=last_hidden_states.detach().numpy()
    #print('feats: ',feats)
    #print('feats.flatten(): ', feats.flatten())
    features=np.array(feats)
    #print('features : ', features)

    # save to file
    return features
```

```python
def split_train_test_transformer():

    import numpy as np


    featsTrain=read_file('/content/drive/MyDrive/GeorgeData/feats_train_transformer.txt')
    featsTest=read_file('/content/drive/MyDrive/GeorgeData/feats_test_transformer.txt')
    list_im_Train=read_file('/content/drive/MyDrive/GeorgeData/train1.txt')
    list_im_Test= read_file('/content/drive/MyDrive/GeorgeData/test1.txt')

    X=[]
    j=0
    #parcourir la liste des images train
    for im_train in list_im_Train:
        #im_train contient le nom de l'image a la position j
        #v_train1 contient le vecteur carateristique relatif à l'image à la position j
        v_train1=featsTrain[j]
        #convertir le vecteur du texte en liste (separateur est le point virgule)
        v_trainspl=v_train1.split(',')
        ##j'affiche la longueur de mon vecteur de caractéristique
        #print('longueur feats  ',len(v_trainspl))
        ##convertir v_trainspl en array numpy
        v_trainx=np.asarray(v_trainspl)
        #forcer le type des valeur en float
        v_train=v_trainx.astype(np.float)
        #je remplis une nouvelle liste X avec les vecteur v_train
        X.append(v_train)
        j+=1

    #convertir la liste X en array
    X=np.array(X)###ensemble train


##le meme travail serait repeté avec l'ensemble de test
    i=0
    Y=[]
    for im in list_im_Test:
        print(im)
        ##read feats test

        v_test1=featsTest[i]
        v_testspl=v_test1.split(',')
        v_testx=np.asarray(v_testspl)
        v_test = v_testx.astype(np.float)
        Y.append(v_test)
        i+=1



    Y=np.array(Y)##ensemble de test (test set)
    return X,Y
```
